```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv("C:\Guvi_projects\Airbnb_collections\Airbnb.csv")
pd.set_option('display.max_columns', None)
```

```python
df.head()
```

```python
x=['name','property_type','room_type','bed_type','bedrooms','beds','number_of_reviews','bathrooms','price','cleaning_fee',
    'host.host_id','host.host_name','host.host_neighbourhood','address.country','review_scores.review_scores_rating']
```

```python
selected_df = df[x]
selected_df.head(5)
```

```python
selected_df.rename(columns={'host.host_id':'host_id','host.host_name':'host_name','host.host_neighbourhood':'host_neighbourhood','address.country':'country','review_scores.review_scores_rating':'review_scores_rating'},inplace=True)
```

```python
selected_df.head(5)
```

```python
selected_df.columns
```

```python
for i in selected_df.columns:
    nulls = selected_df[i].isna().sum()
    print(i,nulls)
```

```python
num_features = ['bedrooms','beds','bathrooms','cleaning_fee','review_scores_rating']
for i in num_features:
    mean = selected_df[i].mean()
    selected_df[i].fillna(mean,inplace=True)
```

```python
cat_features = ['name','host_neighbourhood']
for i in cat_features:
    non_null_values = selected_df[i].dropna().unique()
    selected_df[i] = selected_df[i].apply(lambda x: np.random.choice(non_null_values) if pd.isnull(x) else x)
```

```python
for i in selected_df.columns:
    nulls = selected_df[i].isna().sum()
    print(i,nulls)
```

```python
selected_df.to_csv("C:\Guvi_projects\Airbnb_collections\simplified_airbnb.csv")
```
